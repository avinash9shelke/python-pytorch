```python

import os
import json
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
openai_api_key = os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
    
MODEL = "gpt-4.1-mini"
openai = OpenAI()
```

```python
system_message = """
You are a helpful assistant for an Indian Railways called IRCTC.
Give short, courteous answers, no more than 1 sentence.
Always be accurate. If you don't know the answer, say so.
"""
```

```python
def chat(message,history):
    history = [{"role":h["role"],"content":h["content"]}for h in history]
    messages = [{"role":"system","content":system_message}] + history + [{"role":"user","content":message}]
    response = openai.chat.completions.create(model=MODEL,messages=messages)
    print (f"Request {messages}")
    return response.choices[0].message.content

gr.ChatInterface(fn=chat, type="messages").launch()
```

```python
def get_available_trains(name=None,source=None, destination=None, date=None, berth=None, price=None):
    """
    Advanced filter for train records.
    Supports any combination of: source, destination, date, berth, and price.
    """
    # Start with the full dataset
    results = train_data
    
    if name:
        results = [t for t in results if t["train-name"].lower() == name.lower()]
    # 1. Source & Destination (Exact match, case-insensitive)
    if source:
        results = [t for t in results if t["source"].lower() == source.lower()]
    if destination:
        results = [t for t in results if t["destination"].lower() == destination.lower()]
        
    # 2. Date (Exact match)
    if date:
        results = [t for t in results if t["Date"].lower() == date.lower()]
        
    # 3. Berth/Class (Exact match)
    if berth:
        results = [t for t in results if t["class"].lower() == berth.lower()]
        
    # 4. Price (Range match - Find everything under a budget)
    if price:
        results = [t for t in results if t["price"] <= price]

    return results
```

```python
train_avaibility_function = {
  "name": "get_available_trains",
  "description": "Retrieve the booking details for a specific train journey including source, destination, and pricing.",
  "parameters": {
    "type": "object",
    "properties": {
      "train-name": {
        "type": "string",
        "description": "The name of the train (e.g., Steel Express)"
      },
      "source": {
        "type": "string",
        "description": "The city where the journey begins"
      },
      "destination": {
        "type": "string",
        "description": "The city where the journey ends"
      },
      "class": {
        "type": "string",
        "enum": ["First", "Second AC", "Third AC", "Sleeper", "CC"],
        "description": "The travel class of the ticket"
      },
      "price": {
        "type": "integer",
        "description": "The cost of the ticket in INR"
      },
      "Date": {
        "type": "string",
        "description": "The scheduled date of travel"
      }
    },
    "additionalProperties": False
  }
}
```

```python
tools = [{"type":"function","function":train_avaibility_function}]

tools
```

#### Getting OpenAI to use our Tool

```python
system_message = f"""
You are a helpful assistant for an Indian Railways called IRCTC.
Who provide the train availability for the given source or destination or price or date or class
If the record found in the available trains then return the result in the below format 
Example : 
Train Name : Pune Express
Source : Pune
Destination: Mumbai
Price: 100 Rs
Running Date : 21th April 2026
other wise return empty 
"""
```

```python
import json


def handle_tool_call(message):
    tool_call = message.tool_calls[0]
    if tool_call.function.name == "get_available_trains":
        arguments = json.loads(tool_call.function.arguments)
        source = arguments.get("source")
        destination = arguments.get("destination")
        train_name = arguments.get("train_name")
        berth = arguments.get("class")
        price = arguments.get("price")
        date = arguments.get("date")

        available_trains = get_available_trains(train_name,source, destination,date,berth,price)
        return {
            "role": "tool",
            "content": json.dumps(available_trains),
            "tool_call_id": tool_call.id,
        }

    raise ValueError(f"Unknown tool: {tool_call.function.name}")
```

```python
import json


def _content_for_openai(content):
    """Gradio may pass str or list-shaped content; OpenAI expects str or typed multimodal parts."""
    if content is None:
        return ""
    if isinstance(content, str):
        return content
    if isinstance(content, list):
        parts = []
        for block in content:
            if isinstance(block, str):
                parts.append({"type": "text", "text": block})
            elif isinstance(block, dict):
                if block.get("type") in ("text", "image_url", "input_audio", "refusal"):
                    parts.append(block)
                elif "text" in block:
                    parts.append({"type": "text", "text": block["text"]})
                else:
                    parts.append({"type": "text", "text": json.dumps(block)})
            else:
                parts.append({"type": "text", "text": str(block)})
        if not parts:
            return ""
        if len(parts) == 1 and parts[0].get("type") == "text" and isinstance(parts[0].get("text"), str):
            return parts[0]["text"]
        return parts
    return str(content)


def chat(message, history):
    history = [
        {"role": h["role"], "content": _content_for_openai(h["content"])}
        for h in history
    ]
    messages = (
        [{"role": "system", "content": system_message}]
        + history
        + [{"role": "user", "content": _content_for_openai(message)}]
    )
    response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)
    
    print(f"response {response}")

    if response.choices[0].finish_reason == "tool_calls":
        assistant_message = response.choices[0].message
        tool_message = handle_tool_call(assistant_message)
        messages.append(assistant_message.model_dump(exclude_none=True))
        messages.append(tool_message)
        response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)

    return response.choices[0].message.content
```
